## P1: write a program to find number of occurrences for each item in list and store that output in a list

```python
lst = [5 , 90 , 3 , -1 , 5 , 3 , 8 , 90 , 8 , -1]  

counts = {} 

for i in lst:
    if i in counts:
        counts[i] += 1
    else:
        counts[i] = 1

output_list = list(counts.items())

print(output_list)
```

## P3: Find Greatest Common divisor between two numbers

```python
x=int(input('enter first number'))  
y=int(input('enter second number'))  
ans=[]  
for i in range(min(x,y), 0, -1):
    if x%i==0 and y%i==0:  
        'current answer is', ans.append(i)  
        break
ans  
```
