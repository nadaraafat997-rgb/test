* Given a list, Replace every positive number by 1, Replace every negative number by 2.

```python
l = [19, 20, 21, 10, -53, -23, 10]

for i in range(len(l)):
    if l[i] > 0:
        l[i] = 1
    elif l[i] < 0:
        l[i] = 2

print(l)
```

### Write a while loop to print all the numbers from 0 to 6 except 3 and 5.

```python
i = 0

while i <= 6:
    if i != 3 and i != 5:
        print(i)
    i += 1
```

### Write a function which converts °C to °F
### Temperature in °C can be converted to °F using this formula: °F = (°C x 9/5) + 32.

```python
def celsius_to_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

print(celsius_to_fahrenheit(25))
```

### Write a function called check-season, it takes a month parameter and returns the season: Autumn, Winter, Spring or Summer.

```python
def check_season(month):
    month = month.lower()
    
    if month in ['september', 'october', 'november']:
        return 'Autumn'
    elif month in ['december', 'january', 'february']:
        return 'Winter'
    elif month in ['march', 'april', 'may']:
        return 'Spring'
    elif month in ['june', 'july', 'august']:
        return 'Summer'
    else:
        return 'Invalid month'

print(check_season('March'))
```

### Write a function named reverse_list. It takes a list as a parameter and it returns the reverse of the list.

```python
def reverse_list(my_list):
    reversed_list = []
    
    for i in range(len(my_list) - 1, -1, -1):
        reversed_list.append(my_list[i])
    
    return reversed_list

print(reverse_list([1, 2, 3, 4, 5]))
```

### Write a function named sum_of_numbers. It takes two parameters min_num and max_num and it adds all the numbers in that range.

```python
def sum_of_numbers(min_num, max_num):
    total = 0
    
    for num in range(min_num, max_num + 1):
        total += num
    
    return total

print(sum_of_numbers(1, 5))
```
