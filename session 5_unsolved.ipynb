# Important python function

### 1-enumerate()

```python
l=['omar','ahmed','mohamed','hossam']
l2=[10,20,30,40]
for i,n in enumerate(l2):
    print(i,n)
```

```python
l2=[10,20,30,40]
for i in range(len(l2)):
    print(i)
```

## lambda functions

```python
def even(x):
    if x%2==0:
        return 'even'
    else:
        return '0dd'
    
x=even(9)
x
```
