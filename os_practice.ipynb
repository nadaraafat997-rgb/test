# 1- OS module

```python
import os
```

### makedirs():reate a folder

```python
os.makedirs("session_6/demo_folder/sub_folder")
```

### listdir(): Shows folder contents

```python
print(os.listdir("files_practice"))
```

### create a file with write:

```python
with open("files_practice/demo_folder/test.txt", "w") as file:
    file.write("Hello from OS demo")
```

### rename():rename a file

```python
os.rename(
    "files_practice/demo_folder/test.txt",
    "files_practice/demo_folder/renamed.txt"
)
```

### remove():delete a file

```python
os.remove("files_practice/demo_folder/renamed.txt")
```

### removedirs():deletes a folder only if empty

```python
os.removedirs("files_practice/demo_folder/sub_folder")
```

# Datetime module

```python
from datetime import datetime
```

## now():gets current date time

```python
now = datetime.now()
print(now)
```

## .year/month/day/hour/minute/second

```python
print("Year:", now.year)
print("Month:", now.month)
print("Day:", now.day)
print("Hour:", now.hour)
print("Minute:", now.minute)
print("Second:", now.second)

```

## how to format the date

```python
formatted_date = now.strftime("%m-%d-%Y")
print(formatted_date)
```
